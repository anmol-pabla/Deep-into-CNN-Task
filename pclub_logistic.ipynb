```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
%matplotlib inline
```

```python
df = pd.read_csv("diabetes2.csv")
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sts = StandardScaler()

df_scaled = df.drop('Outcome',axis=1)
df_scaled = pd.DataFrame(sts.fit_transform(df_scaled))


df_scaled['Outcome'] = df.Outcome
df_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
X = df_scaled
x = df_scaled.drop('Outcome',axis=1)
y = df_scaled[['Outcome']]
X_train, X_test, x_train, x_test, y_train, y_test = train_test_split(X, x, y, test_size=0.2, random_state=42)
```

```python
W0 = 0
W1 = 0
W2 = 0
W3 = 0
W4 = 0
W5 = 0
W6 = 0
W7 = 0
W8 = 0


a = 0.001


MSE = np.array([])
correctness = np.array([])
epochs = np.array([])


def sigmoid(output):
    z = 1/(1+math.exp(-output)) 
    return z



for epoch in range(101):
    
    
    count=0
    correct=0
    preds = np.array([])
    pred_exps = np.array([])
    error = np.array([])
    error_x1 = np.array([])
    error_x2 = np.array([])
    error_x3 = np.array([])
    error_x4 = np.array([])
    error_x5 = np.array([])
    error_x6 = np.array([])
    error_x7 = np.array([])
    error_x8 = np.array([])
   
    
    
    for row in x_train.itertuples():

        pred = W0 + W1*row[1]+ W2*row[2] + W3*row[3] + W4*row[4]+ W5*row[5] + W6*row[6] + W7*row[7]+ W8*row[8]
        
        preds = np.append(preds, pred)
        
    
        pred_exp = sigmoid(pred)
        pred_exps = np.append(pred_exps, pred_exp)   

        if pred_exp>0.5: score = 1
        else: score = 0

        score = (score - X_train['Outcome'][row[0]])

        count+=1
        if(score==0) :
          correct+=1
     

    error = pred_exps - X_train['Outcome']

    error_x1 = error*x_train[0]
    error_x2 = error*x_train[1]
    error_x3 = error*x_train[2]
    error_x4 = error*x_train[3]
    error_x5 = error*x_train[4]
    error_x6 = error*x_train[5]
    error_x7 = error*x_train[6]
    error_x8 = error*x_train[7]
    

    MSE_val = (error*error).mean()
    
    if epoch>0:
      epochs = np.append(epochs,epoch)
      print('Epoch ', epoch,": " )

      MSE = np.append(MSE,MSE_val)
      correctness = np.append(correctness,correct/count)
      print("(Loss:",MSE_val,", Correctness:",correct/count,")")
 

    W0 = W0 - a*np.sum(error)
    W1 = W1 - a*np.sum(error_x1)
    W2 = W2 - a*np.sum(error_x2)
    W3 = W3 - a*np.sum(error_x3)
    W4 = W4 - a*np.sum(error_x4)
    W5 = W5 - a*np.sum(error_x5)
    W6 = W6 - a*np.sum(error_x6)
    W7 = W7 - a*np.sum(error_x7)
    W8 = W8 - a*np.sum(error_x8)
```

```python
plt.plot(epochs,correctness)
plt.title("EPOCHS VS CORRECTNESS")
plt.show()
plt.plot(epochs,MSE)
plt.title("EPOCHS VS LOSS")
plt.show()
```

```python
count=0
correct=0
for row in x_test.itertuples():
        
     
        pred = W0 + W1*row[1]+ W2*row[2] + W3*row[3] + W4*row[4]+ W5*row[5] + W6*row[6] + W7*row[7]+ W8*row[8]
        pred_exp = sigmoid(pred)  

        score = (pred_exp - X_test['Outcome'][row[0]])
        
        if pred_exp>0.5: score = 1
        else: score = 0

        score = (score - X_test['Outcome'][row[0]])

        count+=1
        if(score==0) :
          correct+=1
     
print("Test accuracy\nCorrect:",correct,"Total:",count," (",correct*100/count,"%)")    
```
